```python
# 대부분의 코드는 for로 대체할 수 있습니다.
# 효율도 매우 좋지 않습니다. 
# 다이나믹 프로그래밍 등 필수적으로 사용되는 곳이 있기도 합니다.
```

```python
# 재귀함수 버전 팩토리얼 
def factorial(n):
    if n == 1:
        return n
    return n * factorial(n-1)

factorial(5)

#                 return 값
# factorial(5) == 5 * factorial(4) == 5 * 24
# factorial(4) == 4 * factorial(3) == 4 * 6
# factorial(3) == 3 * factorial(2) == 3 * 2
# factorial(2) == 2 * factorial(1) == 2 * 1
# factorial(1) == 1
```

```python
# for문 버전 팩토리얼
result = 1
for i in range(1, 6) :
    result *= i
result
```

```python
# 재귀함수 버전 문자열 반대
def string_reverse(s) :
    if len(s) == 1 :   
        return s
    return string_reverse(s[1:]) + s[0]
string_reverse('hello')

# string_reverse('hello') == string_reverse('ello') + 'h' == 'olle' + 'h' == 'olleh'
# string_reverse('ello') == string_reverse('llo') + 'e' == 'oll' + 'e' == 'olle'
# string_reverse('llo') == string_reverse('lo') + 'l' == 'ol' + 'l' == 'oll'
# string_reverse('lo') == string_reverse('o') + 'l' == 'o' + 'l' == 'ol'
# string_reverse('o') == 'o'
```

```python
# for문 버전 문자열 반대
s = 'hello'
result = ''
for i in s:
    print('-----')
    print(f'i :{i}')
    print(f'연산전 result : {result}')

    result = i + result
    print(f'연산후 result : {result}')

result
```

```python
# 재귀함수 피보나치 수열 
# 재귀가 얼마나 비효율적인지 보여주는 예시.
def fib(num):
    if num == 0:
        return 0
    if num == 1 or num == 2:
        return 1
    else:
        return fib(num-1) + fib(num-2)

fib(35)
```

```python

def plus(num1, num2):
    return num1 + num2

def minus(num1, num2):
    return num1 - num2 

def multiply(num1, num2):
    return num1 * num2

def divide(num1, num2):
    return num1 / num2

print('plus : ',plus(10,5))
print('minus : ',minus(10,5))
print('multiply : ',multiply(10,5))
print('divide : ',divide(10,5))
```

```python
def plus(num1, num2):
    try:
        result = int(num1) + int(num2)
        return result
    except:
        return '연산할 수 없습니다.'

print(f'plus : {plus(10, 5)}')
print(f'plus : {plus("10", "5")}')
print(f'plus : {plus("10", 5)}')
print(f'plus : {plus(10, "5")}')
print(f'plus : {plus("abc", 5)}')
print(f'plus : {plus(10, "abc")}')
```

```python
def plus(num1, num2):
    try:
        result = int(num1) + int(num2)
        return result
    except:
        return float("inf")

print(f'plus : {plus(10, 5)}')
print(f'plus : {plus("10", "5")}')
print(f'plus : {plus("10", 5)}')
print(f'plus : {plus(10, "5")}')
print(f'plus : {plus("abc", 5)}')
print(f'plus : {plus(10, "abc")}')
float("inf") > 1000000000000000000000
```

```python
# lambda version

plus = lambda x, y: x + y
minus = lambda x, y: x - y
multiply = lambda x, y: x * y
divide = lambda x, y: x / y

print(f'plus : {plus(10, 5)}')
print(f'minus : {minus(10, 5)}')
print(f'multiply : {multiply(10, 5)}')
print(f'divide : {divide(10, 5)}')
```

```python
# x의 n제곱을 구하는 함수를 만들어주세요. 재귀함수를 이용하여 만들어야하며, x는 0이 아닙니다.
def power(x, n) :
    if n == 0 :
        return 1
    return power(x, n-1) * x

power(2, 1)
```

```python
# factory 함수
# 클로저
def factory(x):
    def 승수(n):
        return x ** n
    return 승수

제곱3 = factory(3)

# 실제 제곱3는 아래와 같습니다.
# def 승수(n):
#     return 3 ** n

제곱4 = factory(4)

제곱3(2)
제곱3(3)
제곱4(2)
제곱4(3)
```

```python
### 암기 ## 람다 (강사님)
l = [[1, 10, 'leehojun'], 
     [20, 30, 'hojun'], 
     [10, 20, 'weniv'], 
     [1, 2, 'likelion'], 
     [55, 11, 'sun']]

# 1. 글자 수 대로 정렬해주세요.
def 함수(x):
    return len(x[2])

print(list(sorted(l, key=함수)))
print(list(sorted(l, key=lambda x:len(x[2]))))

# 2. 맨 앞에 위치한 숫자대로 정렬해주세요.
def 함수2(x):
    return x[0]

print(list(sorted(l)))
print(list(sorted(l, key=함수2)))
print(list(sorted(l, key=lambda x:x[0])))

# 3. 중앙에 위치한 값대로 정렬해주세요.
def 함수3(x):
    return x[1]

print(list(sorted(l)))
print(list(sorted(l, key=함수3)))
print(list(sorted(l, key=lambda x:x[1])))
```
