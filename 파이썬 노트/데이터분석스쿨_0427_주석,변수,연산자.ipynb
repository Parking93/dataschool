## 암호해독


```python
암호 ='''   + +-+ -+-  
  ++ -- +-+  
  ++-+ -+ -  
  + ++-+ -+   '''

print(암호.split('\n')) 
s = []
for i in 암호.split('\n') :
  s.append(i.replace(' ',''))
l =[]
for i in s :
  l.append(i.replace('+','1').replace('-','0'))
m = []
for i in l :
  m.append(int(i,2))
n = []
for i in m :
  n.append(chr(i))
k = ''
for i in n :
  k += i
print(k)
```

```python
#데이터 분석에서 쓰일만한 데코레이터
def 숫자로변환(func):
    def new_func(l):
        print(func(list(map(int, l))))
    return new_func

@숫자로변환
def 평균(l):
    return sum(l) / len(l)

평균([1, 2, 3, 4, 5])
평균(['1', '2', '3', 4, 5])
평균(['1', '2', '3', '4', '5'])
```
