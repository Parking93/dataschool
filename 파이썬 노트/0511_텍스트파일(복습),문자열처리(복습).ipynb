### 문자열 공백 처리

* strip함수를 사용하여 공백을 제거할 수 있습니다.
* 앞뒤의 공백을 제거할 수 있습니다.

```python
# 문자열 공백처리 예제 

text = ' abcsdefd '
stripped = text.strip()
print('*'+ stripped +'*')
```

```python
text = ' abcsdefd '
print('*'+ text +'*')
```

```python
text = ' abcdefg '

#오른쪽 공백 제거 
r_stripped = text.rstrip()
print(r_stripped)

#왼쪽 공백 제거 
l_stripped = text.lstrip()
print(l_stripped)
```

### 특정 문자열 포함 행만 추출하기

문자열의 split과 리스트 컴프리핸션을 사용하면 특정 문자열을 포함하는 행만 추출 가능

* split을 사용하여 텍스트를 줄바꿈, 1행씩 리스트로 얻을 수 있음
* 리스트 컴프리핸션을 사용하여 특정 문자열을 포함하는 행 추출
* 원래 텍스트로부터 특정한 문자열을 포함하는 행만 추출한 텍스트를 만들 수 있다.

```python
# 미션 - "데이터"가 있는 행을 찾고 싶다.
text = """저는 김철수 입니다.
        반갑습니다. 저는 데이터 분석가 입니다.
        데이터로 일하는 것은 즐거운 일인데
        더 열심히 배우겟습니다. 감사합니다. """

test1 = [s.strip()  for s in text.split('\n') if '데이터' in s]


print(test1)
```

### 텍스트의 빈 행 삭제하기


```python
text = """저는 김철수 입니다.

        반갑습니다. 저는 데이터 분석가 입니다.

        데이터로 일하는 것은 즐거운 일인데
        
        더 열심히 배우겟습니다. 감사합니다. """
text = text.split('\n')

import re

text = [i.strip() for i in text if re.search('[가-힣]', i)]
text = '\n'.join(text)
print(text)
```

```python
text = """저는 김철수 입니다.

        반갑습니다. 저는 데이터 분석가 입니다.

        데이터로 일하는 것은 즐거운 일인데
        
        더 열심히 배우겟습니다. 감사합니다. """
lines = text.split("\n")
line_list = [line.strip() for line in lines if line.strip() != ""] # = 같다 != 다르다. 
new_text = '\n'.join(line_list)
print(new_text)
print(type(new_text))
```
