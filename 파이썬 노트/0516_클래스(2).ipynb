```python
#예제 1 - 틀을 만들자!!

class Dog:
  species = 'firstdog'

  #초기화/인스턴스 속성

  def __init__(self, name, age):
    self.name = name
    self.age = age

# 클래스 정보
print(Dog) 

# 인스턴스화 - 설계화가 된것 생성
a = Dog("mikky", 2)
b = Dog("baby", 3)


# 클래스 변수 - 직접 접근 가능
# 인스턴스 변수 - 객체마다 별도 존재

# 인스턴스 속성 확인
print(f'{a.name} is {a.age} and {b.name} is {b.age}')

# 클래스 변수 - 직접 접근 가능
if a.species == "firstdog":
  print(f'{a.name} is a {a.species}')

# 이런 형태로도 접근 가능하다. 클래스로도 접근 / 인스턴스화로도 접근가능

print(Dog.species)
print(a.species)
print(b.species)

# 사실 이부분이 이해가 되면 클래스가 이해가 된것

# 나만의 공간, 나만의 속성, 공유하는 공간, 공유하는 속성에 대해 이해가 되었기를..
```

```python
def add(num1, num2):
    return num1 + num2

def subtract(num1, num2):
    return num1 - num2

def multiply(num1, num2):
    return num1 * num2

def divide(num1, num2):
    if num2 != 0:
        return num1 / num2
    else:
        print("0으로 나눌 수 없습니다.")

# 사용자로부터 두 개의 숫자 입력 받기
number1 = float(input("첫 번째 숫자를 입력하세요: "))
number2 = float(input("두 번째 숫자를 입력하세요: "))

# 덧셈
result = add(number1, number2)
print("덧셈 결과:", result)

# 뺄셈
result = subtract(number1, number2)
print("뺄셈 결과:", result)

# 곱셈
result = multiply(number1, number2)
print("곱셈 결과:", result)

# 나눗셈
divide(number1, number2)
```

```python
#아래에 요구사항에 맞는 파이썬 클래스 예제를 제시합니다:

class Operation:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        return self.num1 + self.num2

    def subtract(self):
        return self.num1 - self.num2

    def multiply(self):
        return self.num1 * self.num2

    def divide(self):
        if self.num2 != 0:
            return self.num1 / self.num2
        else:
            print("0으로 나눌 수 없습니다.")

# 사용자로부터 두 개의 숫자 입력 받기
number1 = float(input("첫 번째 숫자를 입력하세요: "))
number2 = float(input("두 번째 숫자를 입력하세요: "))

# Operation 클래스의 인스턴스 생성
operation = Operation(number1, number2)

# 덧셈 결과 출력
add_result = operation.add()
print("덧셈 결과:", add_result)

# 뺄셈 결과 출력
subtract_result = operation.subtract()
print("뺄셈 결과:", subtract_result)

# 곱셈 결과 출력
multiply_result = operation.multiply()
print("곱셈 결과:", multiply_result)

# 나눗셈 결과 출력
operation.divide()
```

```python
class pokamon_select_update:
  def __init__(self, name, hp, damage):
    self.name = name
    self.hp = hp
    self.damage = damage

  def attack(self, who):
    print(who, "를 공격합니다.", self.damage, "만큼 데미지를 입혔습니다.") #self에 영향을 받는게 있으면 self를 사용하고, 변수명에도 self를 적용시킨다.

pokamon_1 = pokamon_select_update("피카츄", 40, 5) # 피카츄만의 인스턴스 변수 정의
pokamon_1.attack("파이리")

# 파이리 를 공격합니다. 5 만큼 데미지를 입혔습니다.
```

### 정적 메서드 (@staticmethod)

```python
## 정적 메서드란?

# 다음과 같이 메서드 위에 @staticmethod 을 붙이고, 매개변수에 self를 지정하지 않는다.
# @는 데코레이터 -> 설명 필요

class Calc:
    @staticmethod
    def add(a, b):
        print(a + b)
 
    @staticmethod
    def mul(a, b):
        print(a * b)
 
Calc.add(10, 20)    # 클래스에서 바로 메서드 호출
Calc.mul(10, 20)    # 클래스에서 바로 메서드 호출

# 정잭 메서드는 self를 받지 않으므로 인스턴스 속성에는 접근할 수 없음
```

### 클래스 메서드 (@classmethod)


```python
## 클래스 메서드란?

# 클래스 메서드는 다음과 같이 메서드 위에 @classmethod를 붙인다. 
# 이때 클래스 매서드에는 cls를 지정합니다. -> cls는 class에서 따옴
```

```python
class Person:
    count = 0    # 클래스 속성
 
    def __init__(self):
        Person.count += 1    # 인스턴스가 만들어질 때
                             # 클래스 속성 count에 1을 더함
 
    @classmethod
    def print_count(cls): # cls = Person
        print('{0}명 생성되었습니다.'.format(cls.count))    # cls로 클래스 속성에 접근

james = Person()
maria = Person()
 
Person.print_count()    # 2명 생성되었습니다.
```

```python
# cls가 사용되는 순간 현재 class의 클래스 속성에 count에 접근할 수 있음
# 그래서 print_count 메서드를 호출하면 ~명이 생성되었습니다. 이 출력
# Person,print_count()가 그 결과

# 클래스 메서드는 정적 메서드 처럼 인스턴스 없이 호출 가능

# 그러나 클래스 메서드는 메서드 안에서 클래스 속성이나 클래스 메서드에 접근할때 사용됨
# cls는 가장 중요한 부분인데 메서드 안에서 현재 클래스의 인스턴스를 만들수도 있음
```

```python
class Person:
    count = 0    # 클래스 속성
 
    def __init__(self):
        Person.count += 1    # 인스턴스가 만들어질 때
                             # 클래스 속성 count에 1을 더함
 
    @classmethod
    def print_count(cls):
        print('{0}명 생성되었습니다.'.format(cls.count))    # cls로 클래스 속성에 접근
     
     
    @classmethod
    def create(cls):
        p = cls()    # cls()로 인스턴스 생성 = Person
        return p
        
 
james = Person()
maria = Person()
Person.print_count()    # 2명 생성되었습니다.
```

```python
Person.create()  # cls()로 p라는 인스턴스 생성 -> 현재 Person 클래스를 불러옴
Person.print_count()    # 3명 생성되었습니다.
```

```python
# 둘다 불러오는 방식은 동일하지만 정적메서드에서는 부모 클래스 속성을 가져오고 
# 클래스 매서드는 cls를 통해 현재클래스 속성을 가져온다.

class Test:
    default = "parent"

    def __init__(self):
        self.data = self.default
    
    @classmethod
    def class_test(cls):
        return cls()
    
    @staticmethod
    def static_test():
        return Test()
    
class ChildTest(Test):
    default = "child"

test1 = ChildTest.class_test() #ChildTest의 클래스를 가져옴
test2 = ChildTest.static_test() #부모 클래스 Test를 가져옴

print(test1.default)
print(test2.default)

#출력 결과
#child
#parent

#나중에 사용방법에 대해서 더 알아보도록 합시다!!!
```

## 비공개 속성 

클래스 바깥에서는 접근할 수 없고 클래스 내에서만 사용하는 속성으로
고객이 중요하게 생각하는 정보 요즘 같이 인터넷뱅킹, 스마트뱅킹, 비트코인지갑 등이 많은 이때에는 
더더욱이 잔액관리가 중요합니다.

이러한 정보를 사용하는 방법은 속성이름 앞에 __속성명 으로 사용하면 됩니다.

```python
# 비공개 속성 실습

class Person:
    def __init__(self, name, age, address, wallet):
        self.name = name
        self.age = age
        self.address = address
        self.__wallet = wallet    # 변수 앞에 __를 붙여서 비공개 속성으로 만듦
 
    def pay(self, amount):
        self.__wallet -= amount   # 비공개 속성은 클래스 안의 메서드에서만 접근할 수 있음
        print(amount, '원을 출금하여', self.__wallet, '원 남았네요.')
 
maria = Person('마리아', 20, '서울시 서초구 반포동', 10000)
maria.pay(3000)
```

## 상속

상속은 어떤 클래스가 다른 클래스의 성질을 물려받는것.
유산을 상속 받는것과 동일하다고 생각하며 됨

어떤 클래스를 만들때 처음부터 다른 클래스로 부터 상속받게 만들면 좀더 쓸만한 클래스를 만들수 있음

* 기능을 물려줌 - base class(기반 클래스) - 부모 클래스, 슈퍼클래스
* 상속을 받아 새롭게 만듬 - derived class(파생 클래스) - 자식 클래스, 서브 클래스

[상속을 사용하는 이유]

* 결국 기존 기능을 재사용하고, 중복되는 기능을 만들지 않아도 되기 때문

```python
# class 기반클래스이름:
#     코드
# class 파생클래스이름(기반클래스이름):
#     코드 
```

```python
class Person(object):
  def __init__(self, name, age):
    self.name = name
    self.age = age

class Korean(Person):
  pass

first_korean = Korean("Dawon",4) #Person은 부모
print(first_korean.name)
print(first_korean.age)

```

```python
# 이 코드는 먼저 Person class를 생성한 것
# Person 클래스는 생성자 __init__() 함수를 만들어 name과 age를 만들었다는 것은 이제 알것이다.

# 여기서 중요한 것은 지금까지는 그러지 않았지만 클래스안에 클래스를 볼수 있다.
##### Korean클래스를 만들면서 Person클래스를 상속받는것이다.

# 그리고 pass를 사용하여 별도의 내용 없이 클래스만 존재하는 것을 만들었다.
# 결국 Korean은 아무런 내용이 없는 클래스인데, Korean(Person)을 사용하게 되면서
# Korean은 Person이 가지고 있는 생성자를 그래도 사용할수 있게 된것이다.
```

```python
# 예제 

class Person(object): # 부모 클래스 선언
  def __init__(self, name, age, gender):
    self.name = name
    self.age = age
    self.gender = gender

  def about_me(self): # 메서드 선언
    print(f"안녕하세요 제 이름은 {self.name}이고, 나이는 {self.age}이고, 성별은 {self.gender}입니다.")

# 클래스를 하나 더 만들어 보자! 상속받을 클래스이다.
class Empolyee(Person):
  def __init__(self,name, age, gender, salary, hire_date):
    super().__init__(name, age, gender)
    self.salary = salary
    self.hire_date = hire_date
  
  def do_work(self):
    print("근무시간 이에요")

  def about_me(self):
    super().about_me()
    print(f"저의 월급은 {self.salary}이고, 제 입사날짜는 {self.hire_date}입니다.")

jaehyeok = Empolyee( 'jaehyeok', '30','male','30000', '6/14')
print(jaehyeok.name)
```

```python
class Person:
    def __init__(self):
        print('Person __init__')
        self.hello = '안녕하세요.'
 
class Student(Person):
    def __init__(self):
        print('Student __init__')
        self.school = '데이터 분석 스쿨 입니다.'
 
james = Student()
print(james.school)
print(james.hello)    # 기반 클래스의 속성을 출력하려고 하면 에러가 발생함

# 왜 오류가 났을까? 위에서는 됐던거 같은데;;
# Person의 __init__메서드가 호출되지 않았기 때문에 
# Student의 안녕하세요도 출력되지 않아 속성도 많들어지지 않음
```

```python
# super().__init__()을 사용하여 부모 클래스의 init을 사용할 수 있다.
class Person:
    def __init__(self):
        print('Person __init__')
        self.hello = '안녕하세요.'
 
class Student(Person):
    def __init__ (self):
        super().__init__()
        print('Student __init__')
        self.school = '데이터 분석 스쿨 입니다.' 
 
james = Student()
print(james.school)
print(james.hello)   
```

3. 객체를 생성할때 등록 가능한 대수 출력
4. 차를 신규 등록하는 register()메서드 생성
5. 신규 등록시 등록 현황을 출력
6. 총 주차가능대수가 초과할 경우 “등록할 수 없습니다.” 메세지 출력

```python

class ParkingManager:
    count = 0
    def __init__(self, capacity):
        self.capacity = capacity 
        print(f'총 {self.capacity}대를 등록할 수 있어요.')


    def register(self):
        ParkingManager.count += 1
        if ParkingManager.count > self.capacity:
            print('등록할 수 없습니다.')
        else:
            print(f'신규 차량이 등록되었습니다. ({ParkingManager.count}/{self.capacity})')

manager = ParkingManager(5)
for i in range(6):
  manager.register()

```

```python
class ParkingManager:
    
    def __init__(self, capacity):
        self.capacity = capacity 
        self.count = 0
        print(f'총 {self.capacity}대를 등록할 수 있어요.')


    def register(self):
        self.count += 1
        if self.count > self.capacity:
            print('등록할 수 없습니다.')
        else:
            print(f'신규 차량이 등록되었습니다. ({self.count}/{self.capacity})')

manager = ParkingManager(5)
for i in range(6):
  manager.register()
```

```python
class House:
  #매출 초기화 : 위치, 건물종류 , 매물종류, 가격, 준공연도
  def __init__(self, location, house_type, deal_type, price, completion_year):
    self.location = location
    self.house_type = house_type
    self.deal_type = deal_type
    self.price = price
    self.completion_year = completion_year

  #매출 정보 표시
  def show_detail(self):
    print(self.location, self.house_type, self.deal_type, self.price, self.completion_year)

houses = []
house1 = House("강남", "아파트", "매매", "10.5억", "2010년")
house2 = House("강남", "빌라", "전세", "12.5억", "2015년")
house3 = House("강남", "빌라", "전세", "12.5억", "2015년")

houses.append(house1)
houses.append(house2)
houses.append(house3)

print(f"총 {len(houses)} 개의 매물이 있습니다.")

for house in houses:
  house.show_detail()
```
