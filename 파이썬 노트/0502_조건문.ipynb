### 문제 1. 대학교 학점은 다음과 같이 되어있음

* A - 90점 이상 
* B - 80점 이상
* C - 60점 이상
* D - 40졈 이상
* F - 40점 이하 
* 결석이 3번인 경우 무조건 F

중첩 조건문 또는 연산자를 활용하여 if문을 작성해보세요

```python
# 1) 학점을 조건으로 어떻게 나눌까?
# 2) else에는 어떤것을 할당할까?
# 3) 값을 어떻게 받을까? 
score = int(input('학점을 입력해주세요. '))
abs = int(input('결석일수를 입력해주세요 '))

if abs >= 3:
    grade = "F"
else:
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 60:
        grade = "C"
    else:
        grade = "D" if score >= 40 else "F"
        
print(grade)  
```

### 문제 5. 태어난 연도를 입력받아 생애주기 받기
2023년 6월 28일 부터 만나이로 변경이 되면서 나이에 혼선이 생기고 있는 가운데
정부 연령기준에 맞춰 자신이 어떤 나이대 인지 찾아보자

https://www.gosims.go.kr/hh/hh001/retrieveLfcySearch.do

* 정부기준으로 영유아는 0~5 / 아동은 6~12 / 청소년 13~18 / 청년 19~29 / 중년 30~49 / 장년 50~64 / 노년 65세 이상 / 연령대 무관으로 나눈다.
* 태어난 연도를 입력 받는다.
* 만나이 적용 날짜를 산출한다.
* if문을 통해 어떤 종류의 생애주기 인지를 출력한다.
* 현재 나이와 만니이를 계산하고, 만나이 기준 생애주기를 출력하세요

```python

from datetime import datetime
import pytz

birth_year = int(input('태어난 연도를 적어주세요. '))
now = datetime.now(pytz.timezone('Asia/Seoul')).year
age = now - birth_year
print(f'2023년 6월 28일부터 만나이로 변경이 되면서 당신의 만 나이는 {age}세 입니다.')
if age <= 5 :
    level = '영유아'
elif age <= 12:
    level = '아동'
elif age <= 18:
    level = '청소년'
elif age <= 29:
    level = '청년'
elif age <= 49:
    level = '중년'
elif age <= 64:
    level = '장년' 
else :
    level = '노년'

print(f'현재 나이는 {age + 1}세입니다.')
print(f'현재 만나이는 만{age}세입니다.')
print(f'만나이 기준 생애주기는 {level}입니다.') 
```
