```python
# 강사님이 짠 코드 
import random

# 매칭횟수(콜 받은 횟수)
matching = 0
# 실제로 탄 고객(탑승횟수)
people = 0
# 택시 이용요금(승객에게 받은 요금)
money = 0
# 총 운행 시간(총 소요시간)
all_time = 0

# 운행은 30만원 이상일때 종료
while money < 300000:
  matching += 1
  time = random.randint(5, 20)

  if 5 <= time <= 15:
    people += 1
    each_money = time * 1000
    print(f'손님 {people}이(가) 탑승했습니다. 소요시간은 {time}분이 걸립니다. 택시요금은 {each_money}입니다')
    money += each_money
    all_time += time
  else:
    print('매칭에 실패했습니다.')
print(f"오늘 매칭 횟수는 {matching}번 이고, 탑승손님 {people}명 이고, 소요시간은 {all_time}분 이고, 총 매출은 {money}입니다.")
```
