## 반복문


* 반복문을 사용하는 이유 -> 업무의 최소화, 빠른 퇴근, 나의 체력 안배

특히 데이터 분석에서는 
* 월별/분기별/연도별 산출할 게 있을 경우
* 컬럼별 데이터 검사 및 특이값 이상치 발견
* Rule Base 최적 상품 추천
* ML/DL모델을 파라미터를 바꿔서 반복 시행할때
* 기타 3번 이상 반복 작업 해야하는 반복되는 일들

에서 사용되게 됩니다.


### for문 


```python
# For 반복문 기초 - 튜플

for x in (1, 2, 3, 4, 5) :
    print(f'{x}번째 입니다.')

print('---------------')

# For 반복문 기초 - 문자열
문자열 = 'hello world'
for i in 문자열:
    print(i)

print('---------------')

# For 반복문 기초 - 리스트
list_a = ['하나', '둘', '셋', '넷', '다섯']
for i in list_a :
    print(i)

print('---------------')
```

```python
# For 반복문 기초 - 딕셔너리
dic_a = {'하나' : 1,
         '둘' : 2,
         '셋' : 3,
         '넷' : 4}
for i in dic_a:
    print(i)
print('---------------')

for a in dic_a.items():
    print(a)

print('---------------')
for a in dic_a.values():
    print(a)
```

```python
# For 반복문 기초 - 딕셔너리

d = {"key1" : 1000 , "key2" : 2000, "key3" : 3000}
for i in d.keys():
  print("출력된 문자는", i, "입니다.")

print('--------------------')

# 반복문 기초 7 - 딕셔너리

d = {"key1" : 1000 , "key2" : 2000, "key3" : 3000}
for i in d.values():
  print("출력된 문자는", i, "입니다.")

print('--------------------')

# 반복문 기초 7 - 딕셔너리

d = {"key1" : 1000 , "key2" : 2000, "key3" : 3000}
for i in d.items():
  print("출력된 문자는", i, "입니다.")
```

### for, else 문


```python
# for, else 예제

for i in range(10):
    print(f'{i} 물고기를 잡았습니다.')
    if i == 5 :
        print('만선입니다. 물고기를 다 잡았습니다.')
        break # 더이상 문장을 실행하지 않고, 종료
else:
    print('아직 여유가 좀 있지만, 물고기가 더 없는 것 같으니 이정도로 만족하고 돌아갑시다.')
print('수고하셨습니다.')
    
```

```python
# for, else 예제
l = [0, 3, 1, 10]
for x in l :
    if x < 0 :
        print('음수 발견!')
        break
else:
    print('음수를 발견하지 못하였습니다.')
```

```python
# 리스트 컴프리핸션
# 예제3)

def sumthingFuncthing(i) :
    if i % 100 == 0:
        return i
    else:
        return 0
old_list = [100, 200, 300, 101, 202, 303]
new_list = [sumthingFuncthing(i) for i in old_list]
print(sum(new_list))
```

```python

nowtime = int(input('현재 시각을 입력해주세요. '))
endtime = 16    
time = 0
if nowtime > endtime:
    print('지금은 현재 Q&A시간 입니다.')
elif nowtime == endtime:
    print('지금부터 Q&A시간 입니다.')
else:
    while True:
        time += 1 
        nowtime += 1
        print(f'{time}시간이지나 {nowtime}시가 되었습니다.')
        if nowtime == endtime:
            print('지금부터 Q&A시간 입니다.')
            break
            
```

### break continue pass

```python
# continue

count_amount= [
    (3, 15000),
    (5, 25000),
    (1, 5000),
    (8, 40000),
    (0, 0),
    (2, 10000)
]

for count, amount in count_amount:
  if count < 1:
    continue
  print(f'구매개수는 {count}, 구매가격은 {amount}')
```

```python
# 반복문 문제 5
# 변수 age의 자료형이 정수형일때 숫자 문자형일때 문자가 출력되는 코드

ages = ['one', 8, 'twenty', 40]

for age in ages:
  if #type(age) == int:
    print(age, '숫자')
  else:
    print(age, '문자')

# 1. type(age) == 'int'
# 2. type(age) ==  int 
# 3. type(age) == 'str'
# 4. type(age) == int
# 5. type(age) == str
```

```python
# while 문으로 구구단 만들기

i = 2
while i < 10:
    k = 1
    while k < 10:
        print(i, " * ", k, " = ", i*k)
        k += 1
    i += 1
```

```python
# 실습1. 1~100사이의 숫자를 맞출때까지 반복하는 while문을 작성해주시면 됩니다.

import random
guess_number = random.randint(1, 100)
user_input = int(input('숫자를 입력해서 값을 맞춰보세요.'))

while user_input != guess_number :
    if user_input > guess_number:
        print('숫자가 큽니다')
    else:
        print('숫자가 작습니다.')
    user_input = int(input('숫자를 입력해서 값을 다시 맞춰보세요.'))
else:
    print('정답입니다. ')
```

```python
# 주어진 리스트에서 연속해서 나오는 수의 합이 10 이상인 첫번째
# 부분 리스트의 시작 인덱스를 출력하는 함수 find_start_index를
# 작성하세요. 만약 10 이상인 부분 리스트가 없을 경우 -1을 반환합니다.


def find_start_index(list):
    total = 0
    for idx, i in enumerate(list):
        total += i
        if total >= 10:
            print(f'인덱스 {idx}까지의 합이 10이상입니다.')
            return idx
            break
    else:
        return(-1)
find_start_index([1,2,3])
```
