## 조건문 

```python
# 1000만원이상 10번이상 방문시 vip
# 200만원이상 5번 방문시 gold
# 50만원이상 3번 방문시 silver
# 10만원이상 방문시 normal

if 금액 >= 1000 and 방문횟수 >= 10 :
    print('VIP')
elif 금액 >= 200 and 방문횟수 >= 5 :
    print('Gold')
elif 금액 >= 50 and 방문횟수 >= 3 :
    print('Silver')
elif 금액 >= 10 :
    print('normal')
else :
    print('기타')
```

```python
if 1: # 1 = True
  print('hello')
print('end')
if 0:
  print('hello')
print('end')
if -1:
  print('hello')
print('end')
    
```

```python
#공백이 있는 문자열
if  ' ':
    print('hello')
print('end')
#공백이 없는 문자열
if  '':
    print('hello')
print('end')
```

```python
if not None: #None은 False와 같은의미라고 생각하시면 됩니다.
    print('hello')
print('end')
if None: 
    print('hello')
print('end')
```

```python
# if, else 문 

x = int(input('비교할 수를 입력하세요.'))
y = int(input('비교할 수를 입력하세요.'))
print(f'x = {x}, y = {y}')

if x > y :
    print('x가 y보다 큽니다.')
else :
    print('x가 y보다 작거나 같습니다.')
```

```python
# if, elif, else

x = int(input('비교할 수를 입력하세요.'))
y = int(input('비교할 수를 입력하세요.'))
print(f'x = {x}, y = {y}')

if x > y :
    print('x가 y보다 큽니다.')
elif x < y :
    print('x가 y보다 작습니다.')
else :
    print('x와 y는 같습니다.')
```

```python
score = 81
money = 0

if score >= 90:
    print('mom : i\'m so happy!')
    money += 1000000
elif score >= 80:
    print('mom : i\'m happy!')
    money += 100000
elif score >= 70 and score < 80:
    print('mom : i\'m so...!')
    money += 10000
elif score >= 60 and score < 70:
    print('mom : i\'m so...!')
    money += 1000
else:
    print('mom : i\'m...!')
print(money)
```
