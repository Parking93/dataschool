## * args **kwargs 사용법



파이썬 함수에서는 인수 개수를 변경할 수 있다.
위치 인수에 대해 첫번째, 두번째는 필수, 세번째 이후는 선택으로 정할 수 있는데, 임의로 개수가 변하는 인수를 가변 길이 인수라 부르고, *를 붙이면 가변 길이 인수가 된다. 지정한 인수가 튜플로 지정이 되고, args라는 변수명을 많이 사용한다.

* 결국 동시에 여러 입력값을 주고 싶을때 사용한다.
* 아무 이름이나 사용해도 되긴한다. *만 붙이면
* Tuple처럼 사용이 가능하다. 굳이 리스트나 튜플로 변경하지 않아도 된다.
* 가변인자는 들어오는 인자의 갯수가 언제나 변할 수 있다.

 * args ** kwagrs를 사용하는 이유

**kwargs는 딕셔너리 형태로 가변인수를 받을 수 있는 방법입니다.

kwargs는 딕셔너리 형태로 받는다는 특징이 있습니다.

args와 kwargs의 응용

앞에는 args형식으로 뒤에는 kwargs형식으로 받게되면 정상적으로 작동합니다. 파라미터를 넣을때 순서를 바꾸거나 혼합하면 에러가 발생됩니다.

```python
# 아래 예제는 파라미터의 값을 선언해서 보내는 형식으로 구성되어 있습니다. 
# 예제 뿐만 아니라 다양하게 실습해 보시기 바랍니다.

def print_args_kwargs(*args, **kwargs):
    print('args:', args)
    for x in args:
        print(x)
    print('kwargs:', kwargs)
    for x in kwargs:
        print(x)
        print(kwargs[x])

inputlist=[100, True, 'leehojun']
inputdic={'score':100, 'name':'leehojun', 'age':'10'}
print_args_kwargs(*inputlist)
print('--------')
print_args_kwargs(**inputdic)
print('--------')
print_args_kwargs(*inputlist, **inputdic)
```

```python
# 어떤 함수가 몇번 호출되었는지 확인 실행할때!!

def outer_function():
  """ 외부 함수 """
  count = 0
  def inner_function():
    """내부 함수"""
    nonlocal count # 이 nonlocal 함수는 저장이 가능한 함수 count값을 저장.
    count += 1   
    print(f"실행횟수가 {count}회 실행되었습니다.")

  return inner_function

func1 = outer_function()

# 함수 실행
func1()
func1()
func1()
```

```python
# 출금하기.
def withdraw(money, balance):
    if balance >= money:
        print(f'{money}원을 출금했습니다. 잔액은 {balance - money}원 입니다.')
        return balance - money
    else:
        print(f'잔액이 부족합니다. 뽑을 수 있는 잔액은 {balance}입니다.')
balance = withdraw(2000, balance)
```

```python
# 수수료 부과하기

def withdraw_other(money, balance):
    commission = 500
    print(f'업무시간외에는 수수료가 필요합니다 {money}를 출금하여 수수료가 {commission} 만큼 발생해서 잔액은 {balance - money - commission}입니다.')
    return balance - money - commission

balance = withdraw_other(1000, balance)
```

### map, filter, reduce 문제

```python
숫자 = [1, 2, 3, 4, 5]
승수 = [2, 2, 2, 3, 3]

#문제
#1. 숫자의 승수를 zip으로 맵핑해서 진행하세요.
#2. 숫자를 승수만큼 승수하여 결과값을 표현해주세요.
#3. 승수한 값이 100이상인 값을 출력하세요.
#4. 승수한 값을 모두 더하세요.
```

```python
#1. 숫자의 승수를 zip으로 맵핑해서 진행하세요.
zip = list(map(lambda x, y: [x, y], 숫자, 승수))
print(zip)
```

```python
# 2. 숫자를 승수만큼 승수하여 결과값을 표현해주세요.
제곱값 = list(map(lambda x: x[0] ** x[1], zip))
list(map(lambda x: x[0] ** x[1] , zip))
```

```python
#3. 승수한 값이 100이상인 값을 출력하세요.
list(filter(lambda x: x > 100, list(map(lambda x: x[0] ** x[1], zip))))
```

```python
#4. 승수한 값을 모두 더하세요.
from functools import reduce
reduce(lambda x, y: x + y, 제곱값)
```

```python
# 가변인자(*args) 만들기

# def 함수이름(*매개변수):
#     코드
```

```python
# 패킹 예시
def print_numbers(*args):
  print(args)
  for arg in args:
    print(arg)
```

```python
x = [10]
print_numbers(*x)
# 같은 상황
x = 10
print_numbers(x)
```

```python
x = [10,1]
print_numbers(x)
```

```python
# 키워드 인자 사용하기 딕셔너리 패킹

# 매개변수에 이름으로 일치시키는 인자를 키워드 인자라고 한다.
# 함수를 호출할 때 인자의 값 뿐만 아니라 
# 그 이름까지 명시적으로 지정해서 전달하는 방식이다. 
# 인자의 순서와 용도를 매번 기억하지 않아도 되서 편리하다.
```

```python
def personal_info(name, age, address):
  print('이름: ', name)
  print('나이: ', age)
  print('주소: ', address)

personal_info(name='홍길동', 
              age=30, 
              address='경기도 시흥시')
```

```python
x = {'name': '홍길동', 
     'age': 30, 
     'address': '경기도 시흥시'}

personal_info(**x) # 변수에 **를 붙이면 딕셔너리 형태가 언패킹된다.
```

```python
# 키워드 가변인자(**kwargs) 만들기

# 키워드 인자와 딕셔너리 언패킹의 개념은 키워드 가변인자함수를 만들때 사용한다.
```

```python
# 자료 형이 딕셔너리 일때

def personal_info(**kwargs):
  print(kwargs)
  for kw, arg in kwargs.items():
    print(kw, ': ', arg, sep='')
```

```python
# 값을 한 개 넣어도 되고, 세 개 넣어도 됩니다.
personal_info(name='홍길동')
```

```python
personal_info(name='홍길동', 
              age=30, 
              address='경기도 시흥시')
```

```python
# name: 홍길동
y = {'name': '홍길동', 'age': 30, 'address': '서울시 용산구 이촌동'}
personal_info(**y)
#name: 홍길동
#age: 30
#address: 서울시 용산구 이촌동
```

```python
# 예제 2.
# 결론적으로 *args 및 *kwargs 를 사용하여 변수를 패킹 또는 언패킹할 수 있습니다.

def arg_printer(*args): # 패킹 * 튜플 형태로 변경
  print(args)

list = [1,4,5]
arg_printer(*list) # 언패킹
```

```python
# 예제 4. 다음과 같이 사용하면 언패킹 되었다가 다시 패킹되는 형태로 변경된다.

def arg_printer(**kwargs):
   print(kwargs)

dct = {'param1':5, 'param2':8}
arg_printer(**dct)
```
