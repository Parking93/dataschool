```python
# 함수 - 숫자형 연습 
# 함수안에 있는 변수는 바깥에서 사용할 수가 없다는 것을 명심!!

def num(a, b): # a, b를 활용하는 함수 num 
    print(a + b) # 위에서 언급한 a, b는 밑에서 사용해 주셔야 합니다.

type(num(3, 7))
```

```python
# 함수 - 문자형 입력받는 경우
def print_with_smile(string):
    print(string + ':)')
print_with_smile('안녕하세요')
```

```python
# 함수 - 연습문제, 세 개의 숫자를 입력받아 가장 큰 수를 풀력하는 print_max 함수를 정의하라.
# 단, if문을 사용해서 수를 비교하세요.

def print_max(a, b, c):
    if a >= b and a >= c:
        print(a)
    elif b >= a and b >= c:
        print(b)
    else:
        print(c)

print_max(53, 54, 53)
```

```python
# 함수 - 예제, 입력된 문자열을 역순으로 출력하는 reverse 함수 

def print_reverse(string):
    print(string[::-1])

print_reverse('python')
```

```python
# 함수 - 예제, 평균을 출력하는 함수

def num_ave(score_list):
    print(sum(score_list)/len(score_list))
```

```python
# 함수 - 연습문제, 하나의 리스트를 입력받아서 짝수만 출력하는 함수를 만들어보세요

def num_even(list_num):
    list_ex = []
    for i in list_num:
        if i % 2 == 0:
            list_ex.append(i)
    print(list_ex)
num_even([1, 3, 2, 10, 12, 11, 15])
```

```python
# 함수 - 딕셔너리값도 받아들임
def print_dic(my_dict):
  print(my_dict)

print_dic({"이름" : "홍길동", "나이" : 30, "직업" : "의적"})
```

```python
# 함수 - 예제, 문자열 하나를 입력받아서 인터넷주소를 반환하는 함수를 만들어라
def make_url(string):
    url = 'www.'+string+'.com'
    return url
make_url('naver')
```

```python
# 함수 - 예제, 문자열을 문자열 하나씩으로 하는 리스트

def make_list(string):
    my_list = []
    for i in string:
        my_list.append(i)
    return my_list

make_list('asdf')
```

```python
# 함수 예제, 문자열 대체

def convert_int(string):
  return int(string.replace(',',''))
# replace ',' -> ''으로 변경

convert_int('123,123')
```

```python
# 함수 예제, 리턴 뒤에 아무것도 없으면 아무것도 출력하지 않는다.
def test_01(a):
  if a == 10:
    return # 아무것도 출력하지 않는다.
  else:
    print("10이 아닙니다.")
  
test_01(10)
test_01(1)
```

```python
# 연습문제 5
def say_myself(name, old, woman = True):
  print(f"나의 이름은 {name}입니다.")
  print(f"나의 나이는 {old}입니다.")
  if woman:
    print("여성입니다.")
  else:
    print("남성입니다.")

say_myself("최주영", "35")

#1. 나의 이름은 최주영입니다.
#   나의 나이는 35입니다.
#   여성입니다.
#2. 나의 이름은 최주영입니다.
#   남성입니다.
#   나이는 35살 입니다.
#3. 나의 이름은 최주영입니다.
#   여성입니다.
#   나이는 35살 입니다.
#4. 나의 이름은 최주영입니다.
#   나의 나이는 35입니다.
#   남자입니다.
#5. 나의 이름은 최주영입니다.
#   여성입니다.
```

```python
# 연습문제 7 실행된 함수가 어떤값을 가질지 맞춰주세요

def get_abbr(data_list):
  result = []

  for x in data_list:
    result.append(x[:3])

  return result

get_abbr(['Seoul', 'Anyang', 'Incheon', 'Jeju'])

#1. ['Seoul', 'Anyang', 'Incheon', 'Jeju']
#2. ['Seoul', 'Anyang', 'Incheon']
#3. ['Seo', 'Any', 'Inc', 'Jej']
#4. ['Seo']
#5. error
```
