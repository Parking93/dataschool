## index 설정 

- index_col=

````python
# 특정 열을 인덱스로 데이터를 불러올 때 지정할 수 있음.
excel_index = pd.read_excel('rawData_sheetplus.xlsx', index_col='연도')
excel_index
````

## 사용할 열만 선택하기

- usecols=

````python
# 너무 많은 데이터가 있으면, 따로 컬럼만 부를 수 있음. 
# 원하는 컬럼만 불러오기.

excel_cols = pd.read_excel('rawData_sheetplus.xlsx', usecols=['연차', '매출'])
excel_cols
````

## 제한할 행 선택하기 

- skiprows=
- skipfooter=

````python
# 제외할 행을 지정하고 그 외 나머지 행을 불러오기.

# skiprows=숫자  숫자 만큼 위에서 행을 제외시킨다.
excel_rows = pd.read_excel('rawData_sheetplus.xlsx',header=None ,skiprows=2)

excel_rows
````

````python
# 제외할 행을 지정하고 그 외 나머지 행을 불러오기.

# skiprows=[인덱스]  인덱스위치에 행을 제외시킨다.
excel_rows = pd.read_excel('rawData_sheetplus.xlsx', skiprows=[2, 4])

excel_rows
````
