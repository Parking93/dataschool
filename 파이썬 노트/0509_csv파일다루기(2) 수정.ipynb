## 데이터 함수 (불러오기, 저장하기, 콤마없애기)



```python
# 데이터 불러오는 함수 
def opencsv(filename):
    f = open(filename, 'r')
    reader = csv.reader(f)
    output = []
    for i in reader:
        output.append(i)
    return output
```

```python
# 데이터안에 숫자 문자열안에 콤마를 없애는 함수 만들기.

def switch(listname):
    for i in listname: # 이중 리스트 안에
        for j in i:
            try:
                i[i.index(j)] = float(re.sub(',', '', j))
            except:
                pass
    return listname
```

```python
# 새로운 파일을 생성하는 함수 

def writecsv(filename, the_list):
    with open(filename, 'w', newline = '') as f:
        writer = csv.writer(f, delimiter = ',')
        writer.writerows(the_list)
```

## 서울 인구 비율 데이터(seoul.csv)

```python
import os

import csv

import re
```

```python
# 경로 지정 

os.chdir('/content')
```

```python
# 데이터를 불러오는 함수 만들기

def opencsv(filename):
    f = open(filename, 'r')
    reader = csv.reader(f)
    output = []
    for i in reader:
        output.append(i)
    return output
```

```python
# csv데이터 불러오기
data = opencsv('seoul.csv')
data
```

```python
# 데이터안에 숫자 문자열안에 콤마를 없애는 함수 만들기.

def switch(listname):
    for i in listname: # 이중 리스트 안에
        for j in i:
            try:
                i[i.index(j)] = float(re.sub(',', '', j))
            except:
                pass
    return listname
```

```python
# 콤마가 처리된 데이터

sample = switch(data)
sample[:5] # 콤마 처리가 잘됨.
```

```python
# 각 구의 외국인 비율 구하기.
# 외국인 숫자/ 한국인 + 외국인 = 외국인 비율
for i in sample:
    foregin = 0
    try:
        foregin = round((i[2]/(i[1]+i[2]+i[3]) * 100), 1)
        print(i[0], foregin)
    except:
        pass
```

```python
# 외국인 비율이 3% 이상인 대상만 추출해보세요. (문제)
for i in sample:
    foregin = 0
    try:
        foregin = round((i[2]/(i[1]+i[2]+i[3]) * 100), 1)
        if foregin >= 3 :
            print(i[0], foregin)
    except:
        pass
```

```python
# 외국인 비율을 추가하고 외국인 비율이 3%이상인 구역만 선별한 새로운 csv파일 생성.

new_file = [['행정구역','한국인', '외국인', 'senior','외국인 비율(%)']] # 1행 제목 
for i in sample:
    foregin = 0
    try:
        foregin = round((i[2]/(i[1]+i[2]+i[3]) * 100), 1)
        if foregin >= 3:
            new_file.append([i[0],i[1],i[2],i[3],foregin])
    except:
        pass
new_file
```

```python
# 새로운 파일을 생성하는 함수 

def writecsv(filename, the_list):
    with open(filename, 'w', newline = '') as f:
        writer = csv.writer(f, delimiter = ',')
        writer.writerows(the_list)
```

```python
# seoul_new.csv에 new_list를 넣어서 저장하자.
writecsv('seoul_new.csv', new_file)
```

```python
# 새로운 csv파일을 생성하기.

writecsv('new_apt_list.csv', new_list)
```

## 평균 기온 데이터(CSV_example2_data.csv), 년도인구 데이터(example_4_each.csv)

```python
import os

import csv

import re 
```

```python
# 경로지정 (생략)
os.chdir('/content')
```

```python
# 데이터 불러오는 함수

def opencsv(filename):
    f = open(filename, 'r')
    reader = csv.reader(f)
    output = []
    for i in reader:
        output.append(i)
    return output
```

```python
# 데이터안에 숫자 문자열안에 콤마를 없애는 함수 만들기.

def switch(listname):
    for i in listname: # 이중 리스트 안에
        for j in i:
            try:
                i[i.index(j)] = float(re.sub(',', '', j))
            except:
                pass
    return listname
```

```python
# 온도 데이터 불러오기
data_temp = opencsv('CSV_example2_data.csv')
data_temp
```

```python
# 1행의 제목을 next()로 빼서 처리할 수 있음.
import csv

f = open('CSV_example2_data.csv')
data = csv.reader(f)
header = next(data)
for i in data:
    print(i)
print(header)
f.close()
```

```python
# 가장 더웠던 날을 출력해보자!
# data_temp
data_temp_header = data_temp[0] # 제목을 따로 저장
data_temp = data_temp[1:] # 제목을 제외한 데이터 저장 

max_temp = 0
for row in data_temp:
    row[4] = float(row[4])
    if max_temp < row[4]: # 만약 지금까지 최고 기온값보다 현재 값의 최고기온 값이 더 크다면
        max_temp = row[4]
        max_data = row[0]
print(max_temp, max_data)
```

```python
# 가장 추웠던 날은?

min_temp = 99

for row in data_temp:
    row[4] = float(row[4]) # 문자형 데이터를 숫자형으로 바꾼다.
    if min_temp > row[4]: 
        min_temp = row[4] # 최저기온
        min_data = row[0] # 최저기온 날짜
print(min_temp, min_data)
```

```python
# 년도인구 데이터 불러오기

data_pop = opencsv('example_4_each.csv')

# 년도 리스트
years = []
# 인구수 리스트
populations = []

for line in data_pop[1:]:
    years.append(int(line[0]))
    populations.append(int(line[1]))

print(years)
print(populations)
```

```python
# 실습, 데이터를 토대로 그래프 그리기.
import matplotlib.pyplot as plt

# 그래프를 그릴 도화지 준비하기

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# x축에 년도, y축에 인구수가 있는 선 그래프 만들기.
plt.plot(years, populations, color = 'green', marker = 'o', linestyle = 'solid')

# 제목 더하기
plt.title('Seoul Population Change')

# y축에 레이블 추가하기
plt.ylabel('10Million')

# 그래프 보이기.
plt.show()
```

```python
# 4월과 5월의 최고 기온 데이터를 비교하는 데이터.
import csv 

data_raw = opencsv('CSV_example2_data.csv')
data_4 = []
data_5 = []
for i in data_raw:
    try:
        month = i[0].split('-')
        if month[1] == '04':
            data_4.append(float(i[4]))
        elif month[1] == '05':
            data_5.append(float(i[4]))
    except:
        pass

print(data_4)
print(data_5)
```
