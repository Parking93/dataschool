### 1. Series(시리즈)

- 인덱스를 지정할 수 있음 



```python
import pandas as pd 
```

```python
s = pd.Series([1, 2, 3, 4]) # 리스트 
print(s[0])
```

```python
s2 = pd.Series({'a': 1, 'b': 2, 'c': 3, 'd': 4}) # 딕셔너리 
print(s2)
```

```python
s3 = pd.Series(5, index = [0, 1, 2]) # 스칼라로 시리즈를 만드는 방법 인덱스의 개수 만큼 스칼라의 개수가 생성 된다. 
print(s3)
```

```python
# 활용 
data = [1, 2, 3, 4, 5]
index = ['a', 'b', 'c', 'd', 'e']
series_data = pd.Series(data, index = index) 
print(series_data)
```

```python
s = pd.Series(data)
print(s)
```

#### 산술연산


```python
series = pd.Series([1, 2, 3, 4, 5])

# 덧셈
print(f'series + 100 : \n{series + 100}')
# 뺄셈
print(f'series - 100 : \n{series - 100}')

# 곱셈
print(f'series * 100 : \n{series * 100}')

# 나눗셈
print(f'series / 100 : \n{series / 100}')

# 제곱
print(f'series ** 100 : \n{series ** 100}')

# 나머지
print(f'series % 100 : \n{series % 100}')

# 몫 
print(f'series // 100 : \n{series // 100}')
```

#### Series 객체에 접근

- Keys 
- values

```python
s2 = pd.Series({'a': 1, 'b': 2, 'c': 3, 'd': 4}) # 딕셔너리 

print(s2.keys())
print(s2.values)
print(s2.index)
```

```python
series = pd.Series([1, 2, 3, 4, 5])
print(series.keys())
print(series.index)
print(series.values)
```

```python
# items
print(s2.items()) # zip -> 순회가 가능, 쌍으로 이루어져있음.
```

```python
for index, value in s2.items():
    print(f'인덱스 : {index}, values : {value}')
```

```python
s = pd.Series([10, 20, 30, 40], index = ['a', 'b', 'c', 'd'])
print(s)
```

```python
print(s.iloc[1]) # iloc: (행에 기반한) 열에 접근 
```

```python
print(s.at['a']) # at loc 비슷, 하나의 값만 가져올 때
```

```python
print(s.iat[0])  # 
```

```python
print(s.get('e')) # 없는 값에 대한 처리 방법 , 없을 때 에러가 나지않고 None을 출력 
```

#### 결측값 (NaN, None)

* isna()
* isnull()


Nan(Not a number) -> float

None -> None

```python
import numpy as np
import pandas as pd
```

```python
data = [1, 2, 3, None]

print(np.array(data))
```

```python
print(pd.Series(data)) 
```

```python
s = pd.Series(data)
print(s.isna())
print(s.isnull())
```

```python
# 빈 값이 총 몇개인지 확인하는 방법 
print(s.isna().sum())
print(s.isnull().sum())
```

```python
print(s.notnull().sum())
```

**결측치제거**

```python
s.dropna()
```

```python
# 결측치를 0으로 치환

s.fillna(0)
```

```python
# 결측치를 1로 치환
s.fillna(1)
```

#### series 함수 

```python
s = pd.Series([1, 3, 5, 7, 9])

print(s.count()) # 카운트 함수 
print(s.min())
print(s.max())
print(s.mean())
print(s.std())
print(s.var())
```
