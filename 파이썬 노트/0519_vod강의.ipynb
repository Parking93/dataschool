```python
import time
import random

lunch = ["된장찌개", "피자", "제육볶음", "짜장면"]

while True:
    print(lunch)
    item = input("음식을 추가 해주세요 : ")
    if(item == "q"):
        break
    else:
        lunch.append(item)


set_lunch = set(lunch)
while True:
    print(set_lunch)
    item = input('음식을 삭제해주세요. : ')
    if item == 'q':
        break
    else:
        set_lunch = set_lunch - set([item])

print(set_lunch, '중에서 선택합니다. ')
print('5')
time.sleep(1)
print('4')
time.sleep(1)
print('3')
time.sleep(1)
print('2')
time.sleep(1)
print('1')
time.sleep(1)

random.choice(list(set_lunch))
```

```python
import time
import random

dinner = ['김밥', '김치찌개','부대찌개']

print('추가하고 싶은 음식이 있으시면 "있어"라고 적어주세요')
print('추가하고 싶은 음식이 없으시다면 "없어"라고 적어주세요.')
yes_or_no = input()

if yes_or_no == '있어':
    print(dinner)
    while True:
        print('추가하고싶으신 음식이 없으시다면 "그만"이라고 적어주세요. ')
        item = input('추가하고싶은 음식을 입력해주세요. : ')
        if item == '그만':
            break
        else:
            dinner.append(item)
        print(f'현재 메뉴입니다. : {dinner}')
        
    while True:
        print(f'현재 메뉴 : {dinner}')
        print('삭제하고싶으신 음식이 없으시다면 "그만"이라고 적어주세요. ')
        item = input('삭제하고싶은 음식을 적어주세요. : ')
        if item == '그만':
            break
        else:
            dinner.remove(item)
    
elif yes_or_no == '없어':
    while True:
        item = input('삭제하고싶은 음식을 적어주세요. : ')
        if item == 'q':
            break
        else:
            dinner.remove(item)
print(f'{dinner}중에 오늘의 저녁이 결정됩니다.')
print('5초 후에 오늘의 저녁이 결정됩니다.')
print('5')
time.sleep(1)
print('4')
time.sleep(1)
print('3')
time.sleep(1)
print('2')
time.sleep(1)
print('1')
time.sleep(1)
random.choice(dinner)
```

```python
# 저녁메뉴 추천해주는 프로그램
import random
import time 

dinner = ['매운탕', '제육볶음','삼겹살','김치찌개', '된장찌개']

while True:
    print(f'현재는 {dinner}')
    item = input('추가하실 음식을 적어주세요. : ')
    if item == 'q':
        break
    else:
        dinner.append(item)

while True:
    print(f'현재는 {dinner}')
    answer = input('삭제하실 음식을 적어주세요 : ')
    if answer == 'q':
        break
    else:
        dinner.remove(answer)
print(f'현재 메뉴는 {dinner}입니다.')
print('5초 후에 메뉴가 결정됩니다.')
print('5')
time.sleep(1)
print('4')
time.sleep(1)
print('3')
time.sleep(1)
print('2')
time.sleep(1)
print('1')
time.sleep(1)
random.choice(dinner)
```
